```python
import cv2

img = cv2.imread('Bday.png')
# cv2.imwrite("newbday.png", img)
x, y = img.shape[:2]
print("height= {}, width = {}".format(x, y))
(R, G, B) = img[277, 600]
print("R= ", + R, "G= ", + G, "B= ", +B)
B = img[250, 250, 0]
print("just the value of B=", + B)

```

```python
# RESIZING THE IMAGE TO ANY DIMENSION
size = cv2.resize(img, (1080, 1920))
cv2.imwrite("newbday.png", img)

```

```python
# PLACING A BOX AROUND ANY PIXEL
rectangle = cv2.rectangle(img, (1500, 900), (600, 400), (255, 0, 0), 2)
cv2.imwrite("newbday.png", img)

```

```python
# Adding text to the screen
text_input = cv2.putText(img, 'Happy Birthday', (600, 600), cv2.FONT_ITALIC, 6, (255, 255, 0))
cv2.imwrite("newbday.png", img)
```
